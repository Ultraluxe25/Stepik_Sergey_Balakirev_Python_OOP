





```python
from string import ascii_lowercase, digits

#    TextInput  PasswordInput
class TextInput:
    CHARS = " " + ascii_lowercase
    CHARS_CORRECT = CHARS + CHARS.upper() + digits
    
    @classmethod
    def check_name(cls, name) -> None:
        if len(name) < 3 or len(name) > 50:
            raise ValueError("  name")
        #    
        if not (set(name) < set(cls.CHARS_CORRECT)):
                raise ValueError("  name")

    def __init__(self, name: str = "", size: int = 10):
        self.name = name
        self.size = size
        TextInput.check_name(self.name)
        
    def get_html(self) -> str:
        return f"<p class='login'>{self.name}: <input type='text' size={self.size} />"
        
    
class PasswordInput:
    CHARS = " " + ascii_lowercase
    CHARS_CORRECT = CHARS + CHARS.upper() + digits
    
    @classmethod
    def check_name(cls, name) -> None:
        if len(name) < 3 or len(name) > 50:
            raise ValueError("  name")
        for char in name:
            if not char in cls.CHARS_CORRECT:
                raise ValueError("  name")

    def __init__(self, name: str = "", size: int = 10):
        self.name = name
        self.size = size
        PasswordInput.check_name(self.name)
        
    def get_html(self) -> str:
        return f"<p class='password'>{self.name}: <input type='text' size={self.size} />"

                
class FormLogin:
    def __init__(self, lgn, psw):
        self.login = lgn
        self.password = psw

    def render_template(self):
        return "\n".join(['<form action="#">', self.login.get_html(), self.password.get_html(), '</form>'])


#    
login = FormLogin(TextInput(""), PasswordInput(""))
html = login.render_template()
```
